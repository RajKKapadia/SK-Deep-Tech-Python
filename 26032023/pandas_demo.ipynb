```python
def mu_func(val):
    if val == 'female':
        return 0
    else:
        return 1
```

```python
data['new_col'] = data['Sex'].apply(mu_func)
```

```python
data
```

```python
def age_to_group(age):

    if age > 45.0:
        return 'Old'
    if age >= 30.0 or age < 45.0:
        return 'Young'
    else:
        return 'Younger'
```

```python
data['age_to_group'] = data['Age'].apply(age_to_group)
```
