```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', 'green', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)
ax.set_title('Fruit counts...')
ax.set_xlabel('Fruits')
ax.set_ylabel('Count')
ax.legend(title='Fruit color')

plt.show()
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(8, 5), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('X')  # Add an x-label to the axes.
ax.set_ylabel('Power of X')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
plt.show()
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(8, 5), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2)
```

```python
x = np.linspace(0, 2, 100) 

fig, ax = plt.subplots(1, 3)
ax[0].plot(x, x, label='linear') 
ax[1].plot(x, x**2, label='quadratic')
ax[2].plot(x, x**3, label='cubic')
ax[0].set_xlabel('X')
ax[0].set_ylabel('X vs X')
ax[0].set_title('Linear graph...')
plt.show()
```

```python
fig, axd = plt.subplot_mosaic([['upleft', 'right'],
                               ['lowleft', 'right']], layout='constrained')
axd['upleft'].set_title('upleft')
axd['lowleft'].set_title('lowleft')
axd['right'].set_title('right')
axd['upleft'].plot([1, 2, 3, 4])
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'sg')
plt.ylabel('some numbers')
plt.grid()
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
type(img)
```

```python
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
fig = plt.figure(figsize=(15, 8))

ax = fig.add_subplot(121)
plt.imshow(img)
ax.set_title('One')

ax = fig.add_subplot(122)
plt.imshow(img)
ax.set_title('Two')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 8))

ax = fig.add_subplot(121)
plt.imshow(img)
ax.set_title('One')

ax = fig.add_subplot(122)
plt.imshow(img)
ax.set_title('Two')
plt.show()
```

```python
plt.figure(figsize=(15, 8))

subplot = 230
num_images = 6

for i in range(num_images):

    plt.subplot(subplot+i+1)
    plt.imshow(img)
    plt.title(f'Image - {i+1}')

plt.suptitle('6 Images in one figure...')
plt.show()
```
